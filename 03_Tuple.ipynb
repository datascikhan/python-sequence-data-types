
# 2️⃣ Tuple


Python me **tuple** ek **ordered collection** hai jo **immutable** hota hai aur **duplicate elements** allow karta hai.  
- Tuple ke elements **different data types** ke bhi ho sakte hain.  
- Tuple **parentheses `( )`** me define ki jati hai aur elements **comma `,` se separate** hote hain.  
- Immutable hone ki wajah se tuple ke elements create hone ke baad **modify, add, ya remove** nahi kiye ja sakte.  

**Key Points:**  
- Ordered → Index ke through access possible  
- Immutable → Elements change nahi kar sakte  
- Allows duplicates → Same value multiple times ho sakti hai  
- Useful for → Fixed data, keys in dictionaries, data integrity

```python
colors = ("red", "green", "blue")
print(colors[1])      # green
print(colors[0:2])    # ('red', 'green')
print(type(colors))
```

```python
# Example 1: Basic tuple
numbers = (1, 2, 3, 4, 5)
print(numbers)
```

```python
# Example 2: Tuple with mixed data types
mixed = (10, "Python", 3.14, True)
print(mixed)
```

```python
# Example 3: Tuple with duplicates
duplicates = (1, 2, 2, 3, 4, 4, 5)
print(duplicates)
```

```python
# Example 4: Accessing elements (indexing)
fruits = ("apple", "banana", "cherry")
print(fruits[0])   # First element
print(fruits[-1])  # Last element
```

```python
# Example 5: Slicing a tuple
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])  # (20, 30, 40)
```

# Common Tuple Operations – Indexing & Slicing


Python me **tuples** immutable hote hain, lekin **ordered** hone ki wajah se unke elements **indexing** aur **slicing** ke through access kiye ja sakte hain.  
- **Indexing:** Specific element ko access karne ke liye `tuple[index]` ka use hota hai.  
- **Slicing:** Tuple ke **subset of elements** ko extract karne ke liye `tuple[start:end:step]` ka use hota hai.  
- Tuple me **modification, addition, deletion** possible nahi hai kyunki ye immutable hai.


## **Indexing**

```python
# Example 1: First element
fruits = ("apple", "banana", "cherry")
print(fruits[0])  # 'apple'
```

```python
# Example 2: Last element using negative index
print(fruits[-1])  # 'cherry'
```

```python
# Example 3: Middle element
numbers = (10, 20, 30, 40)
print(numbers[2])  # 30
```

```python
# Example 4: Second last element
letters = ('a', 'b', 'c', 'd', 'e')
print(letters[-2])  # 'd'
```

# Tuple **Slicing**

```python
# Example 1: First two elements
fruits = ("apple", "banana", "cherry", "date")
slice1 = fruits[0:2]  # ('apple', 'banana')
print("Slice 1 (first two elements):", slice1)
```

```python
# Example 2: From index 1 to end
numbers = (10, 20, 30, 40, 50)
slice2 = numbers[1:]  # (20, 30, 40, 50)
print("Slice 2 (index 1 to end):", slice2)
```

```python
# Example 3: Slice with step
letters = ('a', 'b', 'c', 'd', 'e')
slice3 = letters[0:5:2]  # ('a', 'c', 'e')
print("Slice 3 (with step 2):", slice3)
```

```python
# Example 4: Last three elements
slice4 = letters[-3:]  # ('c', 'd', 'e')
print("Slice 4 (last three elements):", slice4)
```
