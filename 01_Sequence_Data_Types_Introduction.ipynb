# Python Sequences Data Types


Python me **sequence data types** wo data types hote hain jisme **ordered collection of items** store kiye jaate hain.  
- Sequence me **elements ka order fix** hota hai aur indexing ke through individual elements access kiye ja sakte hain.  
- Python me common sequence types ye hain:  

1. **List** → Mutable, ordered collection of items, duplicates allowed, defined using `[ ]`  
2. **Tuple** → Immutable, ordered collection of items, duplicates allowed, defined using `( )`  
3. **Range** → Immutable sequence of numbers, often used in loops, defined using `range()`  
4. **String** → Ordered sequence of characters, immutable, defined using `' '` or `" "`  

**Key Features:**  
- Ordered → Elements ka fix order  
- Indexable → Indexing aur slicing possible  
- Iterable → Loops me use kiya ja sakta hai  
- Supports membership testing → `in` keyword
