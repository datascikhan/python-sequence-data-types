# Common String Operations


Python me strings ke sath **multiple operations** perform kiye ja sakte hain jaise indexing, slicing, concatenation, repetition, aur methods.  
- Ye operations strings ko manipulate aur access karne ke liye use hote hain.  
- Strings **immutable** hone ki wajah se original string modify nahi hoti, operations ya methods **new string** return karte hain.

---


## **1. Indexing & Slicing**

**Indexing:**  
- Specific character ko access karne ke liye use hoti hai.  
- Syntax: `string[index]`  
- Positive index ‚Üí start se count, negative index ‚Üí end se count



```python
s = "Python"

# Example 1: First character
print(s[0])  # 'P'
```

```python
# Example 2: Last character
print(s[-1])  # 'n'
```

```python
# Example 3: Third character
print(s[2])  # 't'
```

```python
# Example 4: Second last character
print(s[-2])  # 'o'
```

```python
# Example 5: Using variable for index
i = 4
print(s[i])  # 'o'
```


## **Slicing:**  
- String ka **subset of characters** extract karne ke liye use hoti hai.  
- Syntax: `string[start:end:step]`  
- `start` ‚Üí inclusive, `end` ‚Üí exclusive, `step` ‚Üí optional  



```python
s = "Python"

# Example 1: First three characters
print(s[0:3])  # 'Pyt'
```

```python
# Example 2: From index 2 to end
print(s[2:])  # 'thon'
```

```python
# Example 3: Slice with step
print(s[0:6:2])  # 'Pto'
```

```python
# Example 4: Last three characters
print(s[-3:])  # 'hon'
```

```python
# Example 5: Full string reversed
print(s[::-1])  # 'nohtyP'
```

## **2. Concatenation (`+`)**

**Explanation:**  
- `+` operator ka use do ya zyada strings ko **combine** karne ke liye hota hai.  
- Original strings modify nahi hoti, nayi string create hoti hai.  

---

```python
# Example 1
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)  # 'Hello World'
```

```python
# Example 2
greeting = "Hi, " + "Alice!"
print(greeting)  # 'Hi, Alice!'
```

```python
# Example 3
a = "Python"
b = "3.10"
print(a + b)  # 'Python3.10'
```

```python
# Example 4
s = "Good" + " " + "Morning"
print(s)  # 'Good Morning'
```

```python
# Example 5
name = "Bob"
msg = "Hello " + name
print(msg)  # 'Hello Bob'
```

## **3. Repetition (`*`)**


- `*` operator ka use string ko **multiple times repeat** karne ke liye hota hai.  
- Original string modify nahi hoti, nayi string create hoti hai.  

---

```python
# Example 1
s = "Hi! "
print(s * 3)  # 'Hi! Hi! Hi! '
```

```python
# Example 2
line = "-" * 10
print(line)  # '----------'
```

```python
# Example 3
emoji = "üòä"
print(emoji * 5)  # 'üòäüòäüòäüòäüòä'
```

```python
# Example 4
word = "Python "
print(word * 2)  # 'Python Python '
```

```python

# Example 5
pattern = "*-*"
print(pattern * 4)  # '*-**-**-**-*'
```

## **4. String Methods**


- Python strings ke liye multiple built-in **methods** available hain jo string ko modify karte hain aur **new string** return karte hain.  
- Common methods:  
  - `.upper()` ‚Üí Uppercase me convert  
  - `.lower()` ‚Üí Lowercase me convert  
  - `.split()` ‚Üí String ko list me split kare  
  - `.join()` ‚Üí Iterable ko string me join kare  
  - `.replace()` ‚Üí Substring replace kare



### **.upper()** ‚Äì Convert to Uppercase

```python

# Example 1
s = "hello"
print(s.upper())  # 'HELLO'
```

```python
# Example 2
text = "Python Programming"
print(text.upper())  # 'PYTHON PROGRAMMING'
```

```python
# Example 3
name = "alice"
print(name.upper())  # 'ALICE'
```

### **.lower()** ‚Äì Convert to Lowercase

```python

# Example 1
s = "HELLO"
print(s.lower())  # 'hello'
```

```python
# Example 2
text = "Python Programming"
print(text.lower())  # 'python programming'
```

```python

# Example 3
name = "ALICE"
print(name.lower())  # 'alice'
```

### **.split()** ‚Äì Split string into list

```python



# Example 1
s = "Hello World"
print(s.split())  # ['Hello', 'World']
```

```python
# Example 2
text = "Python,Java,C++"
print(text.split(','))  # ['Python', 'Java', 'C++']
```

```python
# Example 3
line = "a-b-c-d"
print(line.split('-'))  # ['a', 'b', 'c', 'd']
```

### **.join()** ‚Äì Join iterable into string

```python

# Example 1
words = ["Hello", "World"]
print(" ".join(words))  # 'Hello World'
```

```python
# Example 2
letters = ['a', 'b', 'c']
print("-".join(letters))  # 'a-b-c'
```

```python
# Example 3
nums = ['1', '2', '3']
print("".join(nums))  # '123'
```

### **.replace()** ‚Äì Replace substring

```python

# Example 1
s = "Hello World"
print(s.replace("World", "Python"))  # 'Hello Python'
```

```python
# Example 2
text = "banana"
print(text.replace("a", "o"))  # 'bonono'
```

```python
# Example 3
line = "2020-01-01"
print(line.replace("-", "/"))  # '2020/01/01'

```
